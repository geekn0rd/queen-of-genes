### Importing dependencies

```python
from algorithm import genetic_algorithm
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.font_manager as fm
```

### Visualizing the solution

```python
def plot_board(solution):
    n = len(solution)
    fig, ax = plt.subplots()
    ax.axis('off')
    for i in range(n):
        for j in range(n):
            if (i + j) % 2 == 0:
                color = 'white'
            else:
                color = "green"
            ax.add_patch(patches.Rectangle((i, j), 1, 1, color=color))
    for i, j in enumerate(solution):
        ax.text(i + 0.5, j + 0.5, 'Q', color='black', fontsize=15, ha='center', va='center', fontproperties=fm.FontProperties(weight='bold') )
    plt.xlim((0, n))
    plt.ylim((0, n))
    ax.set_aspect('equal', adjustable='box')
    plt.show()
```

### Demo for N = 16

```python
board_size = 16
best_chromosome, msg, gen = genetic_algorithm(board_size, max_gen=1000, pop_size=600)
print(best_chromosome, msg, gen, "steps")
```

```python
plot_board(best_chromosome)
```

### Effect of size on runtime of the algorithm

```python
import timeit

from matplotlib import pyplot as plt

start_time = timeit.default_timer()

dt = []
for n in range(8, 17): 
    best_chromosome, msg, gen = genetic_algorithm(n, max_gen=1000, pop_size=400)
    while(gen < 0):
        best_chromosome, msg, gen = genetic_algorithm(n, max_gen=1000, pop_size=400)
    dt.append(timeit.default_timer()-start_time)
plt.plot(range(8, 17), dt)
plt.xlabel("board size")
plt.ylabel("time(sec)")
```
